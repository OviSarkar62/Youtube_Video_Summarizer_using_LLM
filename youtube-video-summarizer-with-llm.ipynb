```python
from youtube_transcript_api import YouTubeTranscriptApi
```

```python
def get_video_id(url_link):
  return url_link.split("watch?v=")[-1]
```

```python
video_id = get_video_id("https://www.youtube.com/watch?v=SaCYkPD4_K0")
```

```python
transcript = YouTubeTranscriptApi.get_transcript(video_id)
```

```python
transcript
```

```python
transcript_joined = " ".join([line['text'] for line in transcript])
```

```python
import openai
```

```python
api_key = "api_key"
```

```python
prompt = f'Summarize the following YouTube transcript using correct grammar, punctuation, and natural sentence structure. Ensure the summary is well-organized into readable paragraphs and ends with a complete sentence and full stop:\n\nTranscript:\n"{transcript_joined}"'
```

```python
from openai import OpenAI

client = OpenAI(
    api_key=api_key
)

completion = client.chat.completions.create(
    model="gpt-4o-mini",
    store=True,
    temperature=0.7,
    max_tokens=512,
    top_p=1,
    frequency_penalty=0,
    presence_penalty=0,
    messages=[
        {"role": "user", "content": prompt}
    ]
)
```

```python
# ✅ Print only the response text
print(completion.choices[0].message.content)
```
